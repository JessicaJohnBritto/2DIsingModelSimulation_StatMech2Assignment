## The simulation

```python
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
#k=1 assumed everywhere
J=0.5 # kT=4J is mean field prediction. In Question T=2
def initial_config(N):
  initconfig = 2*np.random.randint(2, size=(N,N))-1
  return initconfig
```

```python
def energy_config(config, su=True):
    shifted_up = np.roll(config, shift=-1, axis=0)
    shifted_down = np.roll(config, shift=1, axis=0)
    shifted_left = np.roll(config, shift=-1, axis=1)
    shifted_right = np.roll(config, shift=1, axis=1)
    
    energy = -0.5*J*(shifted_up + shifted_down + shifted_left + shifted_right) * config
    if su:
        energy=np.sum(energy)
    return energy
```

```python
def mcstep(config,beta):
    N=len(config)
    for i in range(N):
        for j in range(N):
            cost = 2*J*config[i,j]*(config[(i+1)%N,j] + config[i,(j+1)%N] + config[(i-1)%N,j] + config[i,(j-1)%N])
            config[i,j]*=-1 if cost<=0 or rand() < np.exp(-cost*beta) else 1
    return config
```

```python
def mcsim(T,N=16):
    n_steps=1000
    n_corr=10
    n_avg=1000
    nt=len(T)
    E,M,C,X = np.zeros(nt), np.zeros(nt), np.zeros(nt), np.zeros(nt)
    n1, n2  = 1.0/(n_avg*N*N), 1.0/(n_avg*n_avg*N*N)
    for ind, t in tqdm(enumerate(T)):     
        E1 = M1 = E2 = M2 = 0
        state = initial_config(N)
        for i in range(n_steps):
            state=mcstep(state,1/t)
        for i in range(n_avg* n_corr):
            state=mcstep(state,1/t)
            if i % n_corr == 0:
                Ene = energy_config(state)
                Mag = np.sum(state)
                E1 = E1 + Ene
                M1 = M1 + Mag
                M2 = M2 + Mag*Mag
                E2 = E2 + Ene*Ene

        E[ind] = n1*E1
        M[ind] = n1*M1
        C[ind] = (n1*E2 - n2*E1*E1)/t/t
        X[ind] = (n1*M2 - n2*M1*M1)/t

    
    return M, E, C, X
```

```python
T=np.linspace(0.01, 4, 100)
M16,E16,C16,X16=mcsim(T)
M25,E25,C25,X25=mcsim(T,25)
M40,E40,C40,X40=mcsim(T,40)
```

## 2 point correlation

Taking Tc at the 28th index

```python
def calculate_G(state, r):
    N = state.shape[0]
    G_values = np.zeros(r)

    for x in range(1, r + 1):
        G_sum = 0.0
        for i in range(N):
            for j in range(N):
                G_sum += state[i, j] * state[(i + x) % N, j]

        G_values[x - 1] = G_sum / (N**2)

    return G_values


def mcsimG(T, r, N=40):
    n_steps=100000
    state = initial_config(N)
    for i in tqdm(range(n_steps)):
        state=mcstep(state,1/T)

    G_values = calculate_G(state, r)
    return G_values
```

```python
Gs = mcsimG(1.34, r=20)
```

```python
Rlogs = np.log(np.linspace(1,20,len(Gs)))
logGs = np.log(np.abs(Gs))
fit_coefficients = np.polyfit(Rlogs, logGs, 1)
fit_line = np.poly1d(fit_coefficients)
plt.scatter(Rlogs, logGs, label='Data Points')
plt.plot(Rlogs, fit_line(Rlogs), color='red', label='Linear Fit')
plt.legend()
plt.title('2-pt correlation')
print(fit_line)
```

$\eta$ is -0.46 but the fit is not that great.
